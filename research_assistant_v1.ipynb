```python
from openai import OpenAI
import os


client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))


def call_llm(prompt, temperature=0.3):
    response = client.chat.completions.create(
    model="gpt-4.1-mini",
        messages=[
            {"role": "system", "content": "You are a precise research assistant."},
            {"role": "user", "content": prompt}
            ],
            temperature=temperature
        )
    return response.choices[0].message.content
```

## Search Topic

This is the only external input of our system

```python
topic = "Impact of interest rates on cryptocurrency markets"
```

## Research Planning (LLM as Planner)

```python
plan_prompt = f"""
Create a concise, structured research plan for the topic below.
Focus on logical research steps.


Topic:
{topic}
"""


research_plan = call_llm(plan_prompt)
research_plan
```

#### Here we need a checkpoint. A human eye to check if the plan is allright and if so, we can continue. If not, we need prompt Enginering for better results.

## Search Query Generation

```python
query_prompt = f"""
Based on the research plan below, generate 5 precise Google search queries.
Return one query per line. No explanations.


Research Plan:
{research_plan}
"""


queries = [q.strip() for q in call_llm(query_prompt).split("\n") if q.strip()]
queries
```

## Web Search Tool

This notebook uses a search API (e.g., SerpAPI).
The agent does not browse â€” it consumes structured search results.

```python
import requests


SEARCH_API_KEY = os.getenv("SEARCH_API_KEY")


def search_web(query, num_results=3):
url = "https://serpapi.com/search"
params = {
"q": query,
"api_key": SEARCH_API_KEY,
"num": num_results
}
response = requests.get(url, params=params)
data = response.json()
return [r["link"] for r in data.get("organic_results", [])]
```
